# Organizing Dotfiles with Git


## Dotfiles

Unix command-line programs are usually configured using plain-text hidden files, such as `.zshrc` where I can store shell customizations,  commonly referred to as **dotfiles**, they are stored in `~`



## Version Control with Git


Method 1: Have your entire home directory under version control (and ignore non-configuration files)
> - **BEWARE!** `git clean` could wipe out everything in your home directory that is not tracked by Git.


Method 2: Keep your configuration files in a separate directory and copy or link them into place (**Recommended**)
> - Impossible to accidentally deleting your other files in `~`
> - Able to track configuration files that belong somewhere other than home direcotry
> - Installing docfiles on new machines is easier
    - `git clone` and then copying or linking files
 


## Install all my `Homebrew` packages using only one command
 

### Try to install all useful things with `Homebrew`:
> 1. Install dependencies
    - For example, to install `pyenv`:
    - Run `brew install pyenv`
> 2. Install GUI MacOS Apps
    - For example, `Atom` app:
    - Run `brew cask install atom`
> 3. Install Mac App Store apps
    - https://github.com/mas-cli/mas
> 4. Install docker images as native commands
    - https://github.com/whalebrew/whalebrew


### Create `Brewfile` file to keep all dependencies/packages

> [Homebrew/homebrew bundle](https://github.com/Homebrew/homebrew-bundle)

Method 1:
```Shell
# go to the root of your project, e.g. `~`
cd ~

# Create the homebrew bundle file `Brewfile`
touch Brewfile

# Add all dependencies into `Brewfile`
open Brewfile #--> read the above doc for details

```

Method 2:
```Shell
# go to the root of your project, e.g. `~`
cd ~

# install brew bundle showing progress
brew bundle -v 

# Create a Brewfile from all the existing Homebrew packages you have installed with
brew bundle dump

```

### Install bundles into a new machine from the `Brewfile`


```Shell
# Having `Brewfile` in the working direcotry, run the following to install all dependencies
brew bundle
```
