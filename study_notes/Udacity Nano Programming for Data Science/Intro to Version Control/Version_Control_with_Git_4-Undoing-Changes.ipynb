# `git commit --amend` - Modify the Last Commit


## Amend commit message

> If your Working Directory is clean i.e. **there aren't any uncommitted changes in the repo**, running `git commit --amend` will let you provide a new commit message. 
- Your code editor will open up and display the original commit message. 
- Just fix a misspelling or completely reword it
- Then save it and close the editor to update the commit message


## Add/Modify Files to commit

Say, you've updated the color of all navigation links. You committed that change and thought you were done. But then you found that a special nav link is not with the color you want. 
- **Method 1**: You could just make a new commit that updates the color for that one link or,
- **Method 2**: what's better,  you can **amend the last commit to include the change**.
> - Edit and save the file(s)
> - Stage the files - `git add .`
> - Include the files in the last commit - `git commit --amend`




# Revert a Commit


## What does `revert` do?

> Git takes the changes that were made in commit and does the exact opposite of them:
- If a character/line is added in commit A, if Git reverts commit A, then Git will make a new commit where the content is deleted
- Vice versa -  if a character/line is removed, then reverting that commit will add that content back


## `git revert` command


```Shell
# revert a specific commit
git revert <SHA>

# revert the last commit
git revert HEAD
```
[`git revert` tutorial](https://www.atlassian.com/git/tutorials/undoing-changes/git-revert)


## Exmple

First, make a change that we will revert later:

```Shell

# Make changes to blog heading, stage and commit it
# check the commits
% git log --oneline

### Commits as follows:
d8e9db3 (HEAD -> master) Set page heading to 'Quests & Crusades'
f70b98f Merge branch 'heading-update'
f9b2ba8 (heading-update) Update heading for test
4753db3 Update heading for Test
9a28adf Merge branch 'sidebar'
6b0eedb (footer) Add footer
ca92941 (sidebar) Add sidebar content
593a4a0 Add side bar
:
```

Second, revert the change:
```Shell
% git convert d8e9db3

# check the result
% git log --oneline

### Commits as follows:
bded20e (HEAD -> master) Revert "Set page heading to 'Quests & Crusades'"
d8e9db3 Set page heading to 'Quests & Crusades'
f70b98f Merge branch 'heading-update'
f9b2ba8 (heading-update) Update heading for test
4753db3 Update heading for Test
9a28adf Merge branch 'sidebar'
6b0eedb (footer) Add footer
ca92941 (sidebar) Add sidebar content
593a4a0 Add side bar
:
```
- Note that a new commit (`bded20e`) is created about the revert.
- Also the heading is reverted to the one before commit `d8e9db3`.


# Reset (Erase) a Commit


## The "Three Trees" of a Repo

- the working directory, 
- the staged snapshot, and 
- the commit history.

<img src="https://wac-cdn.atlassian.com/dam/jcr:0c5257d5-ff01-4014-af12-faf2aec53cc3/01.svg?cdnVersion=1006">


## Mixed, Soft and Hard Resets

A `git reset` takes a specified commit and resets the "three trees" to match the state of the repository at that specified commit with the following three flags:

|Flag|Action on Commit|Impact on File Content in the Working Directory|
|:---|:---|:---|
|`git reset --soft`|Move committed changes to the staging index|Changes unaffected|
|`git reset --mixed` *(default)|unstage committed changes (the unstaged commits become "orphaned" and will be thrown away by Git|Changes unaffected|
|`git reset --hard`|Erase commited changes (commits)|Changes will be erased too|
|


### `git reflog` to access erased commits

> Git does keep track of everything for **about 30 days** before it completely erases anything. To access this content, you'll need to use the `git reflog` command.

# Relative Commit Reference

## Different Commit References

You can reference commits by their
- SHA, 
- tags, 
- branches,
- the special HEAD pointer
- **Ancestry Refs** or **Relative Refs**(special characters to indicate relative reference to the current commit)


## Relative Refs - `^` and `~`

> Both `^` and `~` can be used to indicate the parent commit 
   

### Relative Refs to `HEAD`

How to represent commits with reference to `HEAD`:

|Using `^`|Using `~`|Commit|
|:---|:---|:---|
|HEAD^|HEAD~|The parent of `HEAD` |
||HEAD~1|The parent of the `HEAD`|
|HEAD^^|HEAD~2| The grandparent of the `HEAD`|
|HEAD^^^|HEAD~3| The great grandparent of the `HEAD`|

*Note that `HEAD` is the pointer to the last commit of the current working branch*.


### The difference between `^` and `~`

The main difference when the current commit is a **merge commit** - a merge commit has two parents. 
> - `~` will always follow the first parent of a commit
> - `^` can specify which one to follow if the current commit has two parents
      - `^` indicates the first parent of the commit
      - `^2` indicates the second parent of the commit

**For example, for the following commit:**

```Shell
* 9ec05ca (HEAD -> master) Revert "Set page heading to "Quests & Crusades""
* db7e87a Set page heading to "Quests & Crusades"
*   796ddb0 Merge branch 'heading-update'
|\  
| * 4c9749e (heading-update) Set page heading to "Crusade"
* | 0c5975a Set page heading to "Quest"
|/  
*   1a56a81 Merge branch 'sidebar'
|\  
| * f69811c (sidebar) Update sidebar with favorite movie
| * e6c65a6 Add new sidebar content
* | e014d91 (footer) Add links to social media
* | 209752a Improve site heading for SEO
* | 3772ab1 Set background color for page
|/  
* 5bfe5e7 Add starting HTML structure
```
The current commit is `9ec05ca`, so that the following refer to:

||Commit|Explanation|
|:---|:---|:---|
|`HEAD^^^2`|`4c9749e`| `HEAD^^` is the grandparent of the current commit, and `^2` indicates it shall be the second grandparent, i.e. **the grandparent on the second branch**|
|`HEAD^^^2^1` or `HEAD^^^2~1` |`0c5975a`| `((HEAD^^)^2)^1`: `HEAD^^` is the grandparent of the current commit (`796ddb0`), and `^2` indicates it shall be the second grandparent (`4c9749e`), then `^1` or the equivalent `~1` indicates the first parent (`0c5975a`) |
|`HEAD~6`|`209752a`|because it's the 6th parent commit of the current commit on the `master` branch.|
||
|`HEAD~4^2`|`f69811c`|`HEAD~4` is the 4th parent commit of the current commit (`1a56a81`, it is a **merge commit**), and then `^2` indicates the second parent of this merge commit (the one that got merged in!)|
