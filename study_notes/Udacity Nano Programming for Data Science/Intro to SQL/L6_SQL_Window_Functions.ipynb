# Intro to Window Functions

*Window function performs a calculation across a set of table rows that are somehow related to the current row. This is comparable to the type of calculation that can be done with an aggregate function. But unlike regular aggregate functions, use of a window function does not cause rows to become grouped into a single output row — the rows retain their separate identities.*

### Window Functions is invoked by `OVER` clause

> - The `OVER` clause has three components: partitioning (`PARTITION BY`), ordering(`ORDER BY`), and framing. 
> - Click [here](https://www.postgresql.org/docs/current/sql-expressions.html#SYNTAX-WINDOW-FUNCTIONS) for syntax details.



# First Window Function


## `SUM() OVER` to get running total

Create a running total of standard_amt_usd (in the orders table) over order time with no date truncation. Your final table should have two columns: one with the amount being added for each new row, and a second with the running total.

```SQL
SELECT  standard_amt_usd,
		SUM(standard_amt_usd) OVER (ORDER BY occurred_at) AS running_total
FROM orders
```

## Running Total with `PARTITION BY`

Modify your query from the previous quiz to include partitions. Still create a running total of standard_amt_usd (in the orders table) over order time, but this time, date truncate occurred_at by year and partition by that same year-truncated occurred_at variable. Your final table should have three columns: One with the amount being added for each row, one for the truncated date, and a final column with the running total within each year.

```SQL
SELECT  standard_amt_usd,
		DATE_TRUNC('year', occurred_at) AS year,
		SUM(standard_amt_usd) OVER (PARTITION BY DATE_TRUNC('year', occurred_at) ORDER BY occurred_at) AS running_total
FROM orders
```

# `ROW_NUMBER` & `RANK`


`ROW_NUMBER` and other functions listed in [Table 9.60](https://www.postgresql.org/docs/current/functions-window.html#FUNCTIONS-WINDOW-TABLE) depend on the sort ordering specified by the `ORDER BY` clause of the associated window definition. Rows that are not distinct when considering only the `ORDER BY` columns are said to be *peers*. 

`ROW_NUMBER()` and `RANK()` do not require you to specify a variable within the parentheses.


## `ROW_NUMBER`

`ROW_NUMBER()` displays number of the current row **within its partition**, counting from 1. It numbers the rows according to the `ORDER BY` part of the window statement. 

```SQL
SELECT id, account_id, total,
	   ROW_NUMBER() OVER(PARTITION BY account_id
                   ORDER BY total DESC) AS row_number
FROM orders
WHERE account_id IN (1041, 1051)
```

## `RANK`

`RANK()` is slightly different from `ROW_NUMBER()`. 

As in the following example, the customer with `account_id = 1041` has two orders (order id `25` and `26`) with the same `total` order qty. In this case, they are given the same rank `3`, whereas `ROW_NUMBER()` in the last case gives them different numbers (`3` and `4`). 

```SQL
/* Ranking Total Paper Ordered by Account */
SELECT id, account_id, total,
	   RANK() OVER(PARTITION BY account_id
                   ORDER BY total DESC) AS total_rank
FROM orders
WHERE account_id IN (1041, 1051)

```

## `DENSE_RANK`

Use the last example of customer with `account_id = 1041` to illustrate, 
- `RANK()` would give the identical rows a rank of `3`, then skip rank `4`, so the next result would be `5`.

- `DENSE_RANK()` would still give the two identical order `total` rows a rank of `3`, but the following row would be `4` — no ranks would be skipped, as follows:

```SQL
SELECT id, account_id, total,
	   DENSE_RANK() OVER(PARTITION BY account_id
                   ORDER BY total DESC) AS total_rank
FROM orders
WHERE account_id = 1041

```

|id	|account_id|	total|	total_rank|
|---|---|---|---|
|4324|	1041|	791	|1|
|4323|	1041|	525	|2|
|25	|1041	|395	|3|
|26	|1041	|395	|3|
|27	|1041	|212	|4|

## How `PATITION BY` and `ORDER BY` work together

```SQL
SELECT id,
       account_id,
       standard_qty,
       DATE_TRUNC('month', occurred_at) AS month,
       /* Rank by order month for each account */
       DENSE_RANK() OVER (PARTITION BY account_id
                          ORDER BY DATE_TRUNC('month', occurred_at)) AS month_dense_rank,
       /* Running total of standard_qty for each account by each month*/                   
       SUM(standard_qty) OVER (PARTITION BY account_id
                               ORDER BY DATE_TRUNC('month', occurred_at)) AS sum_standard_qty,
       /* Accmumulated count of order for each account by each month */
       COUNT(standard_qty) OVER (PARTITION BY account_id
                               ORDER BY DATE_TRUNC('month', occurred_at)) AS count_standard_qty,
       /* Average standard_qty per order for each account by each month*/
       AVG(standard_qty) OVER (PARTITION BY account_id
                               ORDER BY DATE_TRUNC('month', occurred_at)) AS avg_standard_qty, 
       /* Smallest standard order for each account by each month */ 
       MIN(standard_qty) OVER (PARTITION BY account_id
                               ORDER BY DATE_TRUNC('month', occurred_at)) AS min_standard_qty,
       /* Largest standard order for each account by each month */                        
       MAX(standard_qty) OVER (PARTITION BY account_id
                               ORDER BY DATE_TRUNC('month', occurred_at)) AS max_standard_qty
FROM orders
WHERE account_id IN (1071, 1081)
```

## Why - Aggregates in Window Functions using `PARTITION` and `ORDER`

The `ORDER BY` clause is one of two clauses integral to window functions. The `ORDER` and `PARTITION` define what is referred to as **the “window” — the ordered subset of data over which calculations are made**. 

Removing `ORDER BY` just leaves an unordered partition; in our query's case, each column's value is simply an aggregation (e.g., sum, count, average, minimum, or maximum) of all the `standard_qty` values in its respective `account_id`.

As Stack Overflow user mathguy explains:
 
> The easiest way to think about this - leaving the `ORDER BY` out is equivalent to "ordering" in a way that all rows in the partition are "equal" to each other. Indeed, you can get the same effect by explicitly adding the `ORDER BY` clause like this: `ORDER BY 0` (or `"order by"` any constant expression), or even, more emphatically, `ORDER BY NULL`.

# Alias for Multiple Window Functions

As we can consider the **ordered subset of data over which calculations are made** as **the window**, we can give it an alias and use it in the query for ease of reading.

For example, we can use alias `main_window` to modify the previous example as belows:
```SQL
SELECT id,
       account_id,
       standard_qty,
       DATE_TRUNC('month', occurred_at) AS month,
       DENSE_RANK() OVER main_window  AS month_dense_rank,
       SUM(standard_qty) OVER main_window AS sum_standard_qty,
       COUNT(standard_qty) OVER main_window AS count_standard_qty,
       AVG(standard_qty) OVER main_window AS avg_standard_qty, 
       MIN(standard_qty) OVER main_window AS min_standard_qty,
       MAX(standard_qty) OVER main_window AS max_standard_qty
FROM orders

WINDOW main_window AS (PARTITION BY account_id
                       ORDER BY DATE_TRUNC('month', occurred_at))
```

### Practice
Use alias to create a window for dataset partitioned by account and ordered by year:

```SQL

SELECT id,
       account_id,
       DATE_TRUNC('year',occurred_at) AS year,
       DENSE_RANK() OVER account_year_window AS dense_rank,
       total_amt_usd,
       SUM(total_amt_usd) OVER account_year_window AS sum_total_amt_usd,
       COUNT(total_amt_usd) OVER account_year_window AS count_total_amt_usd,
       AVG(total_amt_usd) OVER account_year_window AS avg_total_amt_usd,
       MIN(total_amt_usd) OVER account_year_window AS min_total_amt_usd,
       MAX(total_amt_usd) OVER account_year_window AS max_total_amt_usd
FROM orders

WINDOW account_year_window AS (PARTITION BY account_id ORDER BY DATE_TRUNC('year',occurred_at))
```

# Window Functions to Compare a Row to Previous and Next rows


## `LAG` and `LEAD`

You can use `LAG` and `LEAD` functions whenever you are trying to compare the values in
- adjacent rows or 
- rows that are offset by a certain number.

## Example

Imagine you're an analyst at Parch & Posey and you want to determine how the current order's total revenue compares to the next order's total revenue for each account.

```SQL
SELECT occurred_at,
       total_amt_usd,
       LEAD(total_amt_usd, 1) OVER ( ORDER BY occurred_at) AS lead,
       LEAD(total_amt_usd, 1) OVER ( ORDER BY occurred_at) - total_amt_usd AS lead_difference
FROM (
SELECT occurred_at, 
       SUM(total_amt_usd) total_amt_usd
    FROM orders
    GROUP BY 1
) sub
LIMIT 10
```

# Window Functions for Quick Statistics

## `NTILE` to get percentiles without `PARTITION`

`ntile(num_buckets)` - `num_bucket` is an integer ranging from 1 to the argument value, dividing the partition as equally as possible.

```SQL
SELECT id,
       account_id, standard_qty,
       NTILE(4) OVER (ORDER BY standard_qty) AS quartile,
       NTILE(5) OVER (ORDER BY standard_qty) AS quintile,
       NTILE(100) OVER (ORDER BY standard_qty) AS percentile
FROM orders
ORDER BY standard_qty DESC
LIMIT 10
```
