# Data Types

We can see data types as either `categorical` or `quantitative`. 

- We could identify `quantitative variables` as either `continuous` or `discrete`. 
- We also found we could identify `categorical variables` as either `ordinal` or `nominal`.


# Categorical Variables

When analyzing categorical variables, we commonly just look at the count or percent of a group that falls into each level of a category. 
- For example, if we had two levels of a dog category: lab and not lab. We might say, 32% of the dogs were lab (percent), or we might say 32 of the 100 dogs I saw were labs (count).




# Quantitative Variables

However, unlike categorial variable, there are 4 main aspects used to describe quantitative variables:

- Measures of Center
- Measures of Spread
- Shape of the Distribution
- Outliers


# Measures of Center

- Means
- Medians
- Modes


# Measures of Spread

- Range
- Interquartile Range
- Standard Deviation ($\sigma$, `z-score` or sigma)
- Variance

# Measures of Shape

- Visualization with Histogram 
- Shapes: 
    - Symmetric (Normal Distribution), right-skewed, left-skewed
    - One mode, Bi-modal or more modes
- Outliers
